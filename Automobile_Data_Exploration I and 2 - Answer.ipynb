```python
# Use IQR method to remove Outlier
Q1 = df.hp.quantile(0.25)
Q3 = df.hp.quantile(0.75)
Q3, Q1
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.hp<lower_limit)|(df.hp>upper_limit)]
```

```python
df_no_outlier = df[(df.hp>lower_limit)&(df.hp<upper_limit)]
df_no_outlier
```
